```python
import spacy
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Your list of strings
strings = df['AC.Type'].to_list()
strings = [x for x in strings if type(x) == str]
strings = [x.split()[0] for x in strings]
strings = [x.split('-')[0] for x in strings]

# Tokenize and get vectors for each string
vectors = [nlp(text).vector for text in strings]

# Compute the cosine similarity matrix
cosine_similarities = np.dot(vectors, np.transpose(vectors))

# Using the elbow method to determine the optimal number of clusters
wcss = []  # Within-cluster sum of squares

for num_clusters in range(1, 51):  # You can adjust the range as needed
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(cosine_similarities)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 4))
plt.plot(range(1, 51), wcss, marker='o', linestyle='-', markersize=6, markerfacecolor='red')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

```

```python
import spacy
import numpy as np
from sklearn.cluster import KMeans

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Your list of strings
strings = df['AC.Type'].to_list()
strings = [x for x in strings if type(x) == str]
strings = [x.split()[0] for x in strings]
strings = [x.split('-')[0] for x in strings]
strings = [x.lower() for x in strings]



# Tokenize and get vectors for each string
vectors = [nlp(text).vector for text in strings]

# Compute the cosine similarity matrix
cosine_similarities = np.dot(vectors, np.transpose(vectors))

# Use K-Means for clustering
num_clusters = 100  # Adjust the number of clusters as needed
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(cosine_similarities)

# Group the strings
grouped_strings = [[] for _ in range(num_clusters)]
for i, label in enumerate(kmeans.labels_):
    grouped_strings[label].append(strings[i])
```
